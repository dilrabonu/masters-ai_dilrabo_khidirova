```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file
data = pd.read_csv('data.csv')

# Plot a histogram for the 'Volume' column
plt.figure(figsize=(10, 6))
plt.hist(data['Volume'], bins=30, edgecolor='black', color='blue')
plt.title('Histogram of Volume', fontsize=16)
plt.xlabel('Volume', fontsize=14)
plt.ylabel('Frequency', fontsize=14)
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()

# Display the histogram
plt.show()
```
